```python
import requests
from datetime import datetime, timedelta
import pandas as pd

def get_bls_data(series_id, start_year, end_year):
    base_url = "https://api.bls.gov/publicAPI/v2/timeseries/data/"

    payload = {
        'seriesid': [series_id],
        'startyear': start_year,
        'endyear': end_year,
    }

    response = requests.post(base_url, json=payload)

    if response.status_code == 200:
        data = response.json()['Results']['series'][0]['data']
        df = pd.DataFrame(data)
        df['value'] = pd.to_numeric(df['value'])
        df['year'] = pd.to_numeric(df['year'])
        return df
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return None

# Example: Employment data for Washington State (nonfarm employment)
series_id = 'SMS53000000000000001'

# Specify the start and end years
end_year = datetime.now().year
start_year = end_year - 3

# Get the data
employment_data = get_bls_data(series_id, start_year, end_year)

# Display the data
if employment_data is not None:
    print(employment_data)
```
